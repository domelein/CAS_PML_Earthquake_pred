```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#feature generated 24.03.2019
earthquake_daten = '''C:/studium/studium/CAS_PML/Projekt_Arbeit/earthquake/Daten/earthquake_data/'''
```

# Load submissions

```python
svr_submission = pd.read_csv(earthquake_daten+'model/svr/submission_srv_rfb_62900_94-train-test-split_model.csv')
catboost_submission = pd.read_csv(earthquake_daten+'model/catboost/submission_catboost_4920_20_second_submission.csv')
```

```python
plt.figure(figsize=(18,8))
plt.title("Vergleich der Predictions")
plt.xlabel('seq_id')
plt.ylabel('predicted time_to_failure')
plt.plot(svr_submission.seg_id,svr_submission.time_to_failure)
plt.plot(catboost_submission.seg_id,catboost_submission.time_to_failure)
plt.legend(loc='best')
plt.show()
```

```python


t = range(0,len(svr_submission.seg_id))
s1 = svr_submission.time_to_failure
s2 = catboost_submission.time_to_failure         
ig, axs = plt.subplots(2, 1)



axs[0].plot(t, s1, t, s2)
axs[0].set_xlabel('seq')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2)
axs[1].set_ylabel('coherence')

plt.show()
```
