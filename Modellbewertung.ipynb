```python
import time

import pandas as pd
import numpy as np
#data visualization
import matplotlib.pyplot as plt

pd.options.display.precision = 10
```

# Baseline Test 

```python
random_forest= {'MAE_train_data': 0.5969,'MAE_CV_Score': 1.581617,'MAE_CV_Score_max' : 1.592723,'MAE_CV_Score_min':1.572425}
gradient_Boost={'MAE_train_data': 1.966,'MAE_CV_Score': 1.990398,'MAE_CV_Score_max' : 1.998332,'MAE_CV_Score_min':1.983396}
cat_Boost=     {'MAE_train_data': 1.8,'MAE_CV_Score': 1.879656,'MAE_CV_Score_max' : 1.889791,'MAE_CV_Score_min':1.872457}
svr=           {'MAE_train_data': 1.869,'MAE_CV_Score': 2.184047 ,'MAE_CV_Score_max' : 2.577163 ,'MAE_CV_Score_min':1.348986}
ffn=           {'MAE_train_data': 1.774,'MAE_CV_Score': 0,'MAE_CV_Score_max' : 0,'MAE_CV_Score_min':0}
```

```python
MAE_train_data=[0.5969,1.966,1.8,1.869,1.774]
MAE_train_test_data=[1.566,1.988,1.858,0,0]
MAE_CV_Score=[1.581617,1.990398,1.879656,2.184047,0]
MAE_CV_Score_max=[1.592723,1.998332,1.889791,2.577163,0]
MAE_CV_Score_min=[1.572425,1.983396,1.872457,1.348986,0]
```

```python
# set width of bar
barWidth = 0.15

# Set position of bar on X axis
r1 = np.arange(len(MAE_train_data))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4= [x + barWidth for x in r3]
r5= [x + barWidth for x in r4]
```

```python
# Make the plot
plt.figure(figsize=(10,5))
plt.bar(r1, MAE_train_data, color='saddlebrown', width=barWidth, edgecolor='white', label='train_data')
plt.bar(r2, MAE_CV_Score, color='seagreen', width=barWidth, edgecolor='white', label='mean_cv_score')
plt.bar(r3, MAE_CV_Score_max, color='darkgreen', width=barWidth, edgecolor='white', label='min_cv_score')
plt.bar(r4, MAE_CV_Score_min, color='olive', width=barWidth, edgecolor='white', label='max_cv_score')
plt.bar(r5, MAE_train_test_data, color='teal', width=barWidth, edgecolor='white', label='train_test')


# Add xticks on the middle of the group bars
plt.title('Baseline Model Test')
plt.xlabel('Machine Learning Modelle', fontweight='bold')
plt.ylabel('Mean Average Error', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(MAE_train_data))], ['Random Forest', 'GradientBoost', 'CatBoost', 'SVR','FFN'])
 
# Create legend & Show graphic
plt.legend(loc=4)
plt.show()
```
